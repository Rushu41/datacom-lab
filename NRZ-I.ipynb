<h1>NRZ-I Encoding and Decoding 
</h1>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bits ='10111001'
```

<h3>Encoding NRZ-I<h3>

```python
encode = ''
x = '1' 
for i in bits:
    if i == '1':
        x = '0' if x == '1' else '1'
        encode += x
    else:
        encode += x
print(encode)
```

<h3>Decoding NRZ-I<h3>

```python
decode = ''
y = '1'
for i in encode:
    if i != y:
       decode += '1'
       y = i
    else:
        decode += '0'

print(decode)
```

```python
encode_arr = np.array([int(i) for i in encode])
decode_arr = np.array([int(i) for i in decode])
len(encode_array),decode_arr
```

<h3>Check if bits and decode are same or not</h3>

```python
decode == bits
```

<h2>Drawing plots<h2>

```python
plt.figure(figsize=(10, 5))
plt.step(range(0, len(encode_arr) + 1), np.append(encode_arr, encode_arr[-1]), where='post', label="NRZ-I")
plt.axhline(y=0.5,linestyle='--',color ='r')
plt.xlabel('Time')
plt.ylabel('Voltage')
plt.title('NRZ-I Encode')
plt.yticks([0,1])
plt.grid(True)
plt.xlim(0, len(encode_arr))  
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.step(range(0, len(decode_arr) + 1), np.append(decode_arr, decode_arr[-1]), where='post', label="NRZ-I")
plt.axhline(y=0.5,linestyle='--',color ='r')
plt.xlabel('Time')
plt.ylabel('Voltage')
plt.title('NRZ-I Decode')
plt.grid(True)
plt.yticks([0,1])
plt.xlim(0, len(decode_arr))  
plt.show()
```
